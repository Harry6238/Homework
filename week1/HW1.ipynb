```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import BarycentricInterpolator

# Define Runge function
def runge(x):
    return 1 / (1 + x**2)

# Interval
xmin, xmax = -5, 5
x_plot = np.linspace(xmin, xmax, 1000)
y_true = runge(x_plot)

# Different numbers of interpolation nodes
node_counts = [5, 10, 15, 20]

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_true, 'k', linewidth=2, label="Runge function")

for n in node_counts:
    # Equally spaced nodes
    x_nodes = np.linspace(xmin, xmax, n)
    y_nodes = runge(x_nodes)

    # Polynomial interpolation (Barycentric)
    interpolator = BarycentricInterpolator(x_nodes, y_nodes)
    y_interp = interpolator(x_plot)

    # Plot
    plt.plot(x_plot, y_interp, label=f"n={n} nodes")
    plt.plot(x_nodes, y_nodes, 'o')

plt.title("Runge Phenomenon with Equidistant Nodes") #
plt.xlabel("x")
plt.ylabel("f(x)")
plt.ylim(-1, 2)
plt.legend()
plt.grid(True)
plt.show()
```
